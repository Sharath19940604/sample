```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
```

```python
url ='https://dublin.ie/whats-on'
```

```python
#Opening connection , grabbing the page
uClient = uReq(url)
page_html = uClient.read()
```

```python
uClient.close()
```

```python
#Parsing
page_soup = soup(page_html, "html.parser")
print(page_soup)
```

```python
# Finding each events
article = page_soup.findAll('article',{"class":"event card "})
print(article)
```

```python
len(article)
```

```python
eve = article[0]
print(eve)
```

```python
data_list = []
data ={}

for eve in article:
    title = eve.h2.text.strip()
    p_tags = eve.findAll('p')
    location = p_tags[0].text.strip()
    Desc =  p_tags[1].text.strip()
    Time = eve.time.text.strip()
    image_url = eve.div["style"]
    data["title"] = title
    data["location"] = location
    data["Desc"] = Desc
    data["Time"] = Time
    data["image_url"] = image_url
    data_list.append(data)

print(data_list)
```

```python
data = json.dumps(data_list)
```
