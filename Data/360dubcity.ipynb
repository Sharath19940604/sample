```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import json
```

```python
url ='https://www.360dublincity.com/events'
```

```python
#Opening connection , grabbing the page
uClient = uReq(url) 
page_html = uClient.read()
```

```python
uClient.close()
```

```python
#Parsing
page_soup = soup(page_html, "html.parser")
print(page_soup)
```

```python
events = page_soup.findAll('div',{"class":"event-item clearfix "})
events[0]
```

```python
print(len(events))
```

```python
data_list1= []


for event in events:
    title = event.h3.a["title"]
    if event.em:
        date = event.em.text
    else:
        continue
    desc = event.div.text.strip()
    data = {}
    data["title"] = title
    data["date"] = date
    data["desc"] = desc
    header = [elem.next for elem in event.find_all('strong')]
    header = header[1:]
#     print(header)
    values = [elem.next.next for elem in event.find_all('strong')]
    values = values[1:]
#     print(values)
    j=0
    for i in header:
        data[i] = values[j]
        j=j+1

    data_list1.append(data)

print(data_list1)  
```

```python
data = json.dumps(data_list1)
print(data)
```
